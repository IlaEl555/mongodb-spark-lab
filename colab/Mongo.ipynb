```python
transactions_collection.insert_one({
    "TransactionID": 1,
    "Montant": 1000,
    "Type": "Credit"
})
```

```python
print(transactions_collection.count_documents({}))
```

```python
pipeline = [
    {
        "$group": {
            "_id": "$TransactionType",
            "totalAmount": {"$sum": "$TransactionAmount"}
        }
    }
]

result = transactions_collection.aggregate(pipeline)

for doc in result:
    print(doc)
```

```python
pipeline_status = [
    {
        "$group": {
            "_id": "$Transaction Status",
            "count": {"$sum": 1}
        }
    }
]

for doc in transactions_collection.aggregate(pipeline_status):
    print(doc)
```

```python
pipeline_avg = [
    {
        "$group": {
            "_id": None,
            "avgAmount": {"$avg": "$Transaction Amount"}
        }
    }
]

for doc in transactions_collection.aggregate(pipeline_avg):
    print(doc)
```

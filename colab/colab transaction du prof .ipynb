```python
pipeline = [
    {
        "$group": {
            "_id": "$Transaction Type",
            "totalAmount": {
                "$sum": { "$toDouble": "$Transaction Amount" }
            }
        }
    }
]

for doc in transactions_collection.aggregate(pipeline):
    print(doc)
```

```python
pipeline_status = [
    {
        "$group": {
            "_id": "$Transaction Status",
            "count": { "$sum": 1 }
        }
    }
]

for doc in transactions_collection.aggregate(pipeline_status):
    print(doc)
```

```python
pipeline_avg = [
    {
        "$group": {
            "_id": "$Transaction Type",
            "avgAmount": {
                "$avg": { "$toDouble": "$Transaction Amount" }
            }
        }
    }
]

for doc in transactions_collection.aggregate(pipeline_avg):
    print(doc)
```

```python
fraud_count = transactions_collection.count_documents(
    {"Fraud Flag": "True"}
)

print("Nombre de transactions frauduleuses :", fraud_count)
```

```python
pipeline_total = [
    {
        "$group": {
            "_id": None,
            "totalAmount": {
                "$sum": { "$toDouble": "$Transaction Amount" }
            }
        }
    }
]

for doc in transactions_collection.aggregate(pipeline_total):
    print("Montant total des transactions :", doc["totalAmount"])
```
